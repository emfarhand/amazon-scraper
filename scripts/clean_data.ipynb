```python
import pandas as pd

path = r'..\data\raw\dt=2025-12-23\amazon_oil.csv'
df = pd.read_csv(path)

df
```

```python
#missing value handling

df = df.fillna(0)
```

```python
#duplicate value handling

df = df.drop_duplicates(subset=['title'], keep='first')
```

```python
#cast to correct value format

df['title'] = df['title'].astype(str)
df['rating'] = df['rating'].astype(float)
```

```python
#cast to correct value format (review_count)

df['review_count'] = df['review_count'].astype(str)

for index, value in df['review_count'].items():
    if value.endswith('K'):
        df.loc[index, 'review_count'] = float(value[:-1])*1000
    else:
        df.loc[index, 'review_count'] = float(value)

df['review_count'] = df['review_count'].astype(int)
```

```python
#normalize text (title)

df['title'] = df['title'].str.lower().str.strip()
```
